1 - KÜTÜPHANE & VERİ YÜKLEME

```python
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
import joblib

X_train = np.load('npy/X_train_multi.npy')
y_train = np.load('npy/y_train_multi.npy')

print(f"X_train shape: {X_train.shape} | y_train shape: {y_train.shape}")
```

2 - MODEL & GRIDSEARCHCV

```python
knn = KNeighborsRegressor()
multi_knn = MultiOutputRegressor(knn)

param_grid = {
    'estimator__n_neighbors': [150],
    'estimator__weights': ['distance'],
    'estimator__p': [2] #Manh / Oklid
}

grid_search = GridSearchCV(
    estimator=multi_knn,
    param_grid=param_grid,
    cv=3,
    scoring='neg_mean_squared_error',
    verbose=2,
    n_jobs=-1
)

grid_search.fit(X_train, y_train)

print("\nEn iyi parametre kombinasyonu:")
print(grid_search.best_params_)
print(f"En iyi CV MSE: {grid_search.best_score_:.4f}")
```

3 - MODEL KAYIT

```python
best_model = grid_search.best_estimator_
joblib.dump(best_model, 'models/131_knn_model.pkl')
print("En iyi model 'models/131_knn_model.pkl' dosyasına kaydedildi.")
```
